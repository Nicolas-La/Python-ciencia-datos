Realizar analisis de matriz correlacion y explicar que variable estan correlacionadas

```python
exams["total"] = exams["math score"] +exams["reading score"]+exams["writing score"]
dfc= exams.drop(columns= "total")
dfc.corr(method='pearson')
f,ax = plt.subplots(figsize=(7, 7))
sns.heatmap(dfc.corr(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)
plt.show()
```

Observando la matriz de correlacion podemos decir que hay una correlacion muy alta entre las puntuaciones de escritura y lectura. La correlacion es poca para la puntuacion de matematicas y escritura, pero se considera que hay un muy fuerte relacion entre estas
